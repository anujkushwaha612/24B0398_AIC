```python
import pdfplumber
import os
```

```python
pdf_path = "rag_sample.pdf"  # Replace with your PDF path

pages_text = []

with pdfplumber.open(pdf_path) as pdf:
    for page_number, page in enumerate(pdf.pages):
        text = page.extract_text()
        if text:
            pages_text.append(text)
        else:
            print(f"⚠️ No text found on page {page_number + 1}")

print(f"Extracted text from {len(pages_text)} pages.")
```

```python
print(pages_text[0])
```

```python
# Join all page texts into one big string
full_text = "\n".join(pages_text)

print(f"Total characters: {len(full_text)}")
```

```python
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt_tab')

sentences = sent_tokenize(full_text)
print(f"Total sentences: {len(sentences)}")
```

### Sentence Chunking with Overlap

`create_chunks()` breaks a long list of sentences into smaller text chunks

```python
def create_chunks(sentences, chunk_size=100, overlap=20):
    chunks = []
    current_chunk = []
    current_length = 0

    for sentence in sentences:
        words = sentence.split()
        sentence_length = len(words)

        if current_length + sentence_length <= chunk_size:
            current_chunk.append(sentence)
            current_length += sentence_length
        else:
            chunks.append(" ".join(current_chunk))

            # Start new chunk with overlap
            if overlap > 0:
                overlap_words = []
                while current_chunk and len(" ".join(overlap_words).split()) < overlap:
                    overlap_words.insert(0, current_chunk.pop())
                current_chunk = overlap_words.copy()
                current_length = len(" ".join(current_chunk).split())
            else:
                current_chunk = []
                current_length = 0

            current_chunk.append(sentence)
            current_length += sentence_length

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks
```

```python
text_chunks = create_chunks(sentences, chunk_size=100, overlap=20)
```

```python
print(f"Total chunks created: {len(text_chunks)}")
print(f"\nExample chunk (chunk 0):\n{text_chunks[0]}")
```

```python
from sentence_transformers import SentenceTransformer

# Load sentence-transformers model
model = SentenceTransformer('all-MiniLM-L6-v2')

embedding = model.encode(text_chunks[0])
print(f"Embedding shape: {embedding.shape}")
```

### Encoding Text Chunks into Embeddings

took the `text_chunks` generated from the earlier `create_chunks()` function and passed them in `model.encode()` method to get vector representations (embeddings) for each chunk.

```python
embeddings = model.encode(text_chunks)

print(f"Total embeddings: {len(embeddings)}")
print(f"Each embedding shape: {embeddings[0].shape}")
```

### Building a FAISS Index for Embedding Search

```python
import faiss
import numpy as np

dimension = embeddings[0].shape[0]

index = faiss.IndexFlatL2(dimension)

embeddings_np = np.array(embeddings).astype('float32')
index.add(embeddings_np)

print(f"Total vectors in index: {index.ntotal}")
```

```python
query = "What is the submission deadline?"
query_embedding = model.encode([query])

# Search index for top 3 matches
top_k = 3
D, I = index.search(np.array(query_embedding).astype('float32'), top_k)

# Print retrieved chunks
for idx, chunk_idx in enumerate(I[0]):
    print(f"\n🔍 Match {idx+1}:\n{text_chunks[chunk_idx]}")
```

```python
print(f"Distances: {D}")
```

```python
retrieved_context = "\n".join([text_chunks[idx] for idx in I[0]])

print("Combined Retrieved Context:\n")
print(retrieved_context)
```
